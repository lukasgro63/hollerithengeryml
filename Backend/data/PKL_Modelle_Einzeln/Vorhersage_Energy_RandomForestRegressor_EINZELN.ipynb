# Model Training


```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
import joblib
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
import numpy as np

# Define datasets with names
datasets = {'df_rdf': df_rdf, 'df_lgr': df_lgr, 'df_knn': df_knn, 'df_gnb': df_gnb, 'df_dt': df_dt}

# Hyperparameter-Optimierung mit GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_samples': [None, 0.5, 0.8]
}

for name, df_temp in datasets.items():
    X = df_temp.drop(['energy_consumed'], axis=1)
    y = df_temp['energy_consumed']

    print(f'Dataset name: {name}')

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

    grid_search = GridSearchCV(RandomForestRegressor(), param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
    grid_search.fit(X_train, y_train)

    best_model = grid_search.best_estimator_
    best_model.fit(X_train, y_train)

    # Modell exportieren
    model_filename = f'best_model_{name}.pkl'
    joblib.dump(best_model, model_filename)

    # Modell testen
    X_test_sample = np.array([12, 34, 340000]).reshape(1, -1)
    X_test_sample = scaler.transform(X_test_sample)
    X_test_sample2 = np.array([12, 34, 12]).reshape(1, -1)
    X_test_sample2 = scaler.transform(X_test_sample2)
    X_test_sample3 = np.array([5, 10, 12]).reshape(1, -1)
    X_test_sample3 = scaler.transform(X_test_sample3)
    X_test_sample4 = np.array([5, 5, 12]).reshape(1, -1)
    X_test_sample4 = scaler.transform(X_test_sample4)
    X_test_sample5 = np.array([12, 34, 20000]).reshape(1, -1)
    X_test_sample5 = scaler.transform(X_test_sample5)
    X_test_sample6 = np.array([12, 34, 60000]).reshape(1, -1)
    X_test_sample6 = scaler.transform(X_test_sample6)
    predictions = best_model.predict(X_test_sample)
    print(f'Modell {name}: Vorhersage für X_test_sample: {predictions}')
    predictions2 = best_model.predict(X_test_sample2)
    print(f'Modell {name}: Vorhersage für X_test_sample2: {predictions2}')
    predictions3 = best_model.predict(X_test_sample3)
    print(f'Modell {name}: Vorhersage für X_test_sample3: {predictions3}')
    predictions4 = best_model.predict(X_test_sample4)
    print(f'Modell {name}: Vorhersage für X_test_sample4: {predictions4}')
    predictions5 = best_model.predict(X_test_sample5)
    print(f'Modell {name}: Vorhersage für X_test_sample5: {predictions5}')
    predictions6 = best_model.predict(X_test_sample6)
    print(f'Modell {name}: Vorhersage für X_test_sample6: {predictions6}')

    # Berechne und printe Metriken
    y_pred = best_model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)

    print(f'R2 Score: {r2}')
    print(f'RMSE: {rmse}')
    print(f'MAE: {mae}')
    print(f'MSE: {mse}\n')
```
